```python
import pandas as pd

df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
df.shape
```

```python
df = df.drop(columns = ['Address','Date', 'Postcode','Lattitude', 'Longtitude', 'YearBuilt'])
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[['Propertycount', 'Distance', 'Bedroom2','Bathroom', 'Car']] = df[['Propertycount', 'Distance', 'Bedroom2','Bathroom', 'Car']].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea'] =df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace =True)
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first =True)
```

```python
df.head()
```

```python
X = df.drop('Price', axis = 1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn import linear_model

lasso_model = linear_model.Lasso(alpha = 50, max_iter =100, tol =0.1)
lasso_model.fit(X_train, y_train)
```

```python
lasso_model.score(X_test,y_test)
```

```python
lasso_model.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
ridge_model.fit(X_train, y_train)
```

```python
ridge_model.score(X_test,y_test)
```

```python
ridge_model.score(X_train,y_train)
```
