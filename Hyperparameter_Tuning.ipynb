```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].apply(lambda x: iris.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(iris.data, iris.target, test_size = 0.2) 
```

```python
from sklearn import svm
```

```python
model = svm.SVC(kernel='rbf', C = 30, gamma ='auto')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svm.SVC(kernel = 'linear', C = 10, gamma = 'auto'), iris.data, iris.target, cv = 5)
```

```python
cross_val_score(svm.SVC(kernel = 'rbf', C = 10, gamma = 'auto'), iris.data, iris.target, cv = 5)
```

```python
cross_val_score(svm.SVC(kernel = 'rbf', C = 150, gamma = 'auto'), iris.data, iris.target, cv = 5)
```

```python
kernals = ['rbf', 'linear']
C = [1,10,20]
avg_score = {}
for kvals in kernals:
    for cvals in C:
        cv_score = cross_val_score(svm.SVC(kernel = kvals, C = cvals, gamma = 'auto'), iris.data, iris.target, cv = 5)
        avg_score[kvals + '_' + str(cvals)] = np.average(cv_score)

avg_score
```

GridsearchCV

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(svm.SVC(gamma= 'auto'), {
    'C' : [1,10,20],
    'kernel' : ['rbf', 'linear']
}, cv = 5, return_train_score = False)

clf.fit(iris.data, iris.target)
clf.cv_results_
```

```python
results =pd.DataFrame(clf.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```
