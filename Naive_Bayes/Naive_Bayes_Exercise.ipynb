```python
#import libraries
import pandas as pd
from sklearn.datasets import load_wine
```

```python
df = load_wine()
```

```python
df
```

```python
dir(df)
```

```python
#Create the dataset into a Pandas Data frame
dataset = pd.DataFrame(df.data, columns = df.feature_names)
dataset.head()
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.data, df.target, test_size = 0.2)    
```

```python
len(X_train)
```

```python
len(X_test)
```

Gaussian Bayes 

```python
from sklearn.naive_bayes import GaussianNB

G_model = GaussianNB()
G_model.fit(X_train,y_train)
```

```python
g_predict = G_model.predict(X_test)
```

```python
G_model.score(X_test,y_test)
```

```python
y_test
```

Multinomial Bayes

```python
from sklearn.naive_bayes import MultinomialNB

M_model = MultinomialNB()
M_model.fit(X_train,y_train)
```

```python
M_model.score(X_test,y_test)
```

```python
m_predict = M_model.predict(X_test)
m_predict
```

```python
y_test
```

Evaluation


Gaussian Bayes

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test,g_predict)
sns.heatmap(cm, annot = True)
plt.xlabel('predicted_values')
plt.ylabel('truth_values')
```

Multinomial

```python
cm = confusion_matrix(y_test,m_predict)
sns.heatmap(cm, annot = True)
plt.xlabel('predicted_values')
plt.ylabel('truth_values')
```
